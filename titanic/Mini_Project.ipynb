````python
import pandas as pd
import numpy as np
df = pd.read_csv(r"titanic.csv")
df.head()
````

A quick note on this version of the dataset: unlike a lot of trimmed-down Titanic samples
you'll see online, this one has the full original columns, including a genuine unique key —
`PassengerId` — plus `Name`, `Ticket`, and `Cabin` in their raw form. A few things that changes
for us:
- **Duplicates become a much more meaningful check.** With a real unique ID per row, any
  `duplicated()` result (checked across all columns) tells you something is *actually* wrong with
  the data — it's not just "these passengers happen to share every visible feature" the way it
  would be without an ID.
- **`Cabin` is mostly missing** (most passengers, especially in lower classes, have no recorded
  cabin) — a different, more extreme missing-data situation than anything in our Employees table.
- **`Name` is real, raw text** — including each passenger's title (`Mr.`, `Mrs.`, `Miss.`,
  `Master.`, and a few rarer ones) embedded right in the string. That's genuinely useful data,
  not just a label, once you pull it apart with `.str` methods.

````python
# Your work starts here.
# Task 1: Explore
print(df.shape)
print(df.info())
print(df.describe())
print(df[["Age","Cabin","Embarked"]].isna().sum())


````

````python
# Task 2: Select
new_df=df[["Survived","Pclass","Age","Sex","Fare"]]

new_df
````

````python
# Task 3: Filter
filter_df=df[(df["Sex"]=="female")&(df["Pclass"]==1)&df["Survived"]==1]
filter_df

````

````python
# Task 4: Sort
sorted_df=df.sort_values("Fare",ascending=False)
print(sorted_df["Name"].head(10))
# sorted_df

````

````python
#Task 5: Create AgeGroup

df["Age"] = df["Age"].fillna(df["Age"].mean())

df["AgeGroup"] = ""

df.loc[df["Age"] < 13, "AgeGroup"] = "Child"
df.loc[df["Age"].between(13, 19), "AgeGroup"] = "Teen"
df.loc[df["Age"].between(20, 59), "AgeGroup"] = "Adult"
df.loc[df["Age"] >= 60, "AgeGroup"] = "Senior"
df["AgeGroup"]

````

````python
# Task 6: Handle missing values
df["Cabin"]=df["Cabin"].fillna(df["Cabin"].mode()[0])
df["Cabin"].isna().sum()
# df["Cabin"] = df["Cabin"].fillna(
#     df.groupby("Pclass")["Cabin"].transform(lambda x: x.mode().iloc[0])
# )
# # df["Cabin"].ffill()
# df.groupby(["Pclass"])["Cabin"].agg(["size"])
````

````python
# Task 7: Remove duplicates
print(df.duplicated().sum())
print(df.duplicated(subset=["PassengerId"]).sum())
##PassengerId is just an identifier 

````

````python
# Task 8: Group and aggregate
print(df["Sex"].value_counts())
print(df.groupby(["Sex"])["Survived"].mean())
print(df.groupby(["Pclass"])["Survived"].mean())
"""
Females had a much higher survival rate than males.
Passengers in higher classes had a higher survival rate.
First-class passengers were the most likely to survive, while third-class passengers had the lowest survival rate.

"""

````

````python
# Task 9: Pivot table
pivot_table=pd.pivot_table(df, values="Survived", index="Pclass", columns="Sex", aggfunc="mean")
pivot_table


````

````python
# Task 10: Transform -- extract Title from Name
df["Title"] = df["Name"].str.extract(r',\s*([^\.]+)\.')
# print(df["Title"].value_counts())
print(df.groupby(["Title"])["Survived"].mean())

````
