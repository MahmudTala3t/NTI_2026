```python
df = df.drop(columns=["ManagerNotes"])
print(df.shape)
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean()).astype("int32")
df["EducationField"]=df["EducationField"].fillna(df["EducationField"].mode()[0])
df["JobRole"]=df["JobRole"].fillna(df["JobRole"].mode()[0])
df["MonthlyIncome"]=df["MonthlyIncome"].fillna(df["MonthlyIncome"].mean())
df["TotalWorkingYears"]=df["TotalWorkingYears"].fillna(df["TotalWorkingYears"].mean())
df["YearsAtCompany"]=df["YearsAtCompany"].fillna(df["YearsAtCompany"].mean())



df.isna().sum()[df.isna().sum() > 0]
```

```python
df.duplicated().sum()
df[df["RecordID"].duplicated()]
```

```python
df.drop_duplicates(df,inplace=True)
df.duplicated().sum()
```

```python
df.drop_duplicates(subset=["RecordID"],inplace=True)
df[df["RecordID"].duplicated()]
```

```python
df.drop(columns=["MaritalStatus","HourlyRate","DailyRate","Over18","YearsWithCurrManager","YearsInCurrentRole","MonthlyRate","EmployeeCount","EmployeeNumber","EnvironmentSatisfaction","EducationField","StandardHours","RelationshipSatisfaction","RecordID","JobRole","PerformanceRating","Gender","PercentSalaryHike","Education"],inplace=True)
df.shape
```

```python
def iqr_bounds(series):
    q1, q3 = series.quantile(0.25), series.quantile(0.75)
    iqr = q3 - q1
    return q1 - 1.5 * iqr, q3 + 1.5 * iqr
print(f"Lower and upper : {iqr_bounds(df["MonthlyIncome"])}")
print(iqr_bounds(df["Age"]))
l,u=iqr_bounds(df["MonthlyIncome"])
low,up=iqr_bounds(df["Age"])
df["MonthlyIncome"] = df["MonthlyIncome"].clip(lower=l, upper=u)
df["Age"] = df["Age"].clip(lower=low, upper=up)
```
