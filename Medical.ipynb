```python
import pandas as pd 
import numpy as np
df=pd.read_csv(r"Medicine.csv")
```

```python
print(df.head())
print(df.tail(2))
print(df["Price"].isna().sum())
print(df["Quantity"].isna().sum())
```

```python
df["Price"]=df["Price"].fillna(df["Price"].mean())
df["Quantity"]=df["Quantity"].fillna(1)
print(df["Price"].isna().sum())
print(df["Quantity"].isna().sum())
```

```python
df["Inventory_Value"]=df["Price"]*df["Quantity"]
df
```

```python
well_stocked=df[df["Quantity"]>=15]
print(well_stocked)
df.loc[5:10,["Medicine","Inventory_Value"]]
```

```python
df["Inventory_Value"].describe()
```

```python
df.to_csv("cleaned_medicine_inventory.csv", index=False)
```
