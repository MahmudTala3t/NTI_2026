## 1. Introduction

**What is Pandas?** A Python library for working with **structured (tabular) data** — rows and columns, like Excel or a SQL table, but programmable.

**Why not just NumPy?** NumPy arrays only hold **one data type** and have no labels. Pandas adds **labeled rows/columns** and lets each column have its **own dtype** — exactly what real-world data (mixed text + numbers) needs.

| | NumPy | Pandas |
|---|---|---|
| Structure | array (matrix) | labeled table |
| Best for | numeric computation | data analysis / cleaning |
| Built on | — | NumPy |


```python
import pandas as pd
import numpy as np

print("pandas version:", pd.__version__)
```

```python
# Quick NumPy vs Pandas contrast
arr = np.array([21, 22, 20])
s = pd.Series([21, 22, 20], index=["Sara", "Ali", "Omar"])

print("NumPy array (no labels):\n", arr)
print("\nPandas Series (labeled):\n", s)
```

## 2. Pandas Data Structures

### 2.1 Series — a single labeled column

```python
ages = pd.Series([21, 22, 20, 23])
print(ages)
print(type(ages))
```

Notice the left column — that's the **index** (auto-generated: 0,1,2,3 by default).

```python
# Custom index -- give each value a meaningful label instead of 0,1,2,3
gpa = pd.Series([3.8, 3.5, 3.9], index=["Sara", "Ali", "Omar"])
print(gpa)

# from a dict -- keys automatically become the index
gpa_dict = pd.Series({"Sara": 3.8, "Ali": 3.5, "Omar": 3.9})
print(gpa_dict)
```

```python
# Accessing elements
print(gpa["Sara"])          # by label
print(gpa.iloc[0])          # by position
print(gpa.iloc[0:2])        # slice by position
print(gpa[["Sara", "Omar"]])  # multiple labels
```

```python
# Series attributes
print("index :", gpa.index)
print("values:", gpa.values)      # returns a NumPy array underneath!
print("dtype :", gpa.dtype)
print("shape :", gpa.shape)
print("size  :", gpa.size)
print("ndim  :", gpa.ndim)
```

### 2.2 DataFrame — a full table

A DataFrame = **rows + columns + column names + an index**. Think of it as a dictionary of Series sharing the same index.

```python
# From a dictionary (most common)
employees_small = pd.DataFrame({
    "Name": ["Sara", "Omar", "Lina"],
    "Department": ["IT", "Sales", "IT"],
    "Salary": [9000, 7000, 8500]
})
employees_small
```

```python
# From a list of lists + explicit column names
rows = [["Sara", "IT", 9000], ["Omar", "Sales", 7000], ["Lina", "IT", 8500]]
df_from_lists = pd.DataFrame(rows, columns=["Name", "Department", "Salary"])
df_from_lists
```

```python
# From a list of dictionaries (one dict = one row)
records = [
    {"Name": "Sara", "Department": "IT", "Salary": 9000},
    {"Name": "Omar", "Department": "Sales", "Salary": 7000},
]
df_from_records = pd.DataFrame(records)
df_from_records
```

```python
# From a NumPy array
arr2d = np.array([[1, 90], [2, 85], [3, 78]])
df_from_np = pd.DataFrame(arr2d, columns=["ID", "Score"])
df_from_np
```

```python
# Custom index on a DataFrame
df_custom_idx = pd.DataFrame(
    {"Salary": [9000, 7000, 8500]},
    index=["Sara", "Omar", "Lina"]
)
df_custom_idx
```

```python
# DataFrame is really a dict of Series sharing the same index
print(type(employees_small["Name"]))
employees_small["Name"]
```

```python
# DataFrame properties
df = employees_small
print("shape  :", df.shape)
print("columns:", df.columns.tolist())
print("index  :", df.index.tolist())
print("dtypes :\n", df.dtypes)
print("size   :", df.size)
print("ndim   :", df.ndim)
```

## 5. Selecting Data

- Single column → `df["col"]` returns a **Series**
- Multiple columns → `df[["col1","col2"]]` returns a **DataFrame**
- `.loc[]` → by **label**
- `.iloc[]` → by **integer position**


```python
print(type(df["Salary"]))
df["Salary"]
```

```python
print(type(df[["Name", "Salary"]]))
df[["Name", "Salary"]]
```

```python
print(df.loc[2, "Salary"])        # row labeled 2, column "Salary"
print(df.iloc[2, 4])                # row at position 2, column at position 4
```

```python
df.iloc[0:3, [1, 4]]        # first 3 rows, columns at positions 1 and 4 (Name, Salary)
```

```python
df.loc[0:3, ["Name", "Department"]]     # loc slicing is INCLUSIVE of the end label
```

## 6. Filtering Data

A condition like `df["Salary"] > 8000` builds a **Boolean mask**; `df[mask]` keeps only the `True` rows.

```python
df[df["Salary"] > 8000]
```

```python
# combining conditions -- use & (and), | (or), ~ (not) -- ALWAYS wrap each condition in ()
df[(df["Department"] == "IT") & (df["YearsExperience"] >= 5)]
```

```python
df[(df["Department"] == "IT") | (df["Department"] == "HR")]
```

```python
df[~(df["Department"] == "IT")]     # NOT IT
```

```python
df[df["Department"].isin(["IT", "HR"])]     # same idea as chained == / or, but cleaner
```

```python
df[df["Salary"].between(7000, 9000)]     # inclusive on both ends
```

## 9. Missing Data

`NaN` ("Not a Number") is Pandas' placeholder for missing/undefined values.

```python
df.isna().sum()      # count of missing values PER COLUMN
```

```python
df[df["PerformanceScore"].isna()]     # show just the affected rows
```

```python
df["PerformanceScore"].notna().sum()   # count of NON-missing values
```

```python
# fillna -- fill with a constant
df["PerformanceScore_filled0"] = df["PerformanceScore"].fillna(0)

# fillna -- fill with the column mean (more realistic for a score)
df["PerformanceScore_filledMean"] = df["PerformanceScore"].fillna(df["PerformanceScore"].mean())

df[["Name", "PerformanceScore", "PerformanceScore_filled0", "PerformanceScore_filledMean"]]
```

```python
# forward fill -- carry the previous valid value forward
sample_series = pd.Series([10, np.nan, np.nan, 20, np.nan])
print(sample_series.ffill())     # modern equivalent of fillna(method="ffill")
```

```python
# dropna -- remove rows/columns with any missing value
print("before dropna:", df.shape)
df_complete_cases = df.dropna(subset=["PerformanceScore"])
print("after dropna (subset=PerformanceScore):", df_complete_cases.shape)

df_no_missing_cols = df.dropna(axis=1)      # drop any COLUMN that has at least one NaN
print("columns remaining after dropna(axis=1):", df_no_missing_cols.columns.tolist())
```
