## Rebuilding `employees` (continuing from Part 1)

```python
import pandas as pd
import numpy as np

employees = pd.DataFrame({
    "EmployeeID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 102],  # 102 repeated on purpose
    "Name": ["Sara", "Omar", "Lina", "Ahmed", "Farida", "Youssef", "Mona", "Karim", "Salma", "Omar"],
    "Department": ["IT", "Sales", "IT", "HR", "Sales", "IT", "HR", "Sales", "IT", "Sales"],
    "City": ["Cairo", "Giza", "Cairo", "Alex", "Cairo", "Giza", "Alex", "Cairo", "Giza", "Giza"],
    "Salary": [9000, 7000, 8500, 6000, 7200, 9500, 6100, 7300, 8800, 7000],
    "YearsExperience": [5, 3, 6, 2, 4, 7, 1, 3, 5, 3],
    "PerformanceScore": [88, 75, np.nan, 65, 90, np.nan, 70, 82, np.nan, 75],
})
employees["YearsExperience"] = employees["YearsExperience"].astype(float)
employees["Bonus"] = employees["Salary"] * 0.1
employees["SalaryK"] = employees["Salary"] / 1000
employees["SalaryAfterRaise"] = employees["Salary"] * 1.05

df = employees
df
```

## 11. Grouping & Aggregation

`groupby()` follows **split → apply → combine**: split rows into groups by a column, apply a calculation to each group, combine results into one table.

```python
df.groupby("Department")["Salary"].mean()
```

**🔮 Predict the Output:** which department do you expect to have the highest average salary, based on eyeballing the table above?

```python
df.groupby("Department")["Salary"].mean().round(2).sort_values(ascending=False)
```

```python
df.groupby("Department")["Salary"].sum()
df.groupby("Department")["Salary"].count()
df.groupby("Department")["Salary"].min()
df.groupby("Department")["Salary"].max()
df.groupby("Department")["Salary"].std()
```

```python
# .agg() -- multiple aggregations at once, and DIFFERENT functions per column
df.groupby("Department").agg({
    "Salary": "mean",
    "YearsExperience": "max",
    "EmployeeID": "count"
})
```

```python
df.groupby("Department")["Salary"].agg(["sum", "mean", "count"])
```

```python
# multiple-column grouping
df.groupby(["Department", "City"])["Salary"].mean()
```

**✍️ Practice 11**

2. For each `Department`, get `min`, `max`, and `mean` of `Salary` in one call using `.agg()`.
3. Group by `["Department", "City"]` and count employees in each group.


```python
# ✅ solution
print(df.groupby("Department")["YearsExperience"].mean())
print(df.groupby("Department")["Salary"].agg(["min", "max", "mean"]))
print(df.groupby(["Department", "City"])["EmployeeID"].count())
```

## 12. Combining DataFrames

```python
print(df)
departments = pd.DataFrame({
    "Department": ["IT", "Sales", "HR", "Finance"],   # Finance has no employees in df, on purpose
    "Manager": ["Nadia", "Hassan", "Rana", "Tarek"],
    "Budget": [500000, 350000, 200000, 150000],
})
departments
```

```python
merged = pd.merge(df, departments, on="Department", how="left")
merged[["Name", "Department", "Manager", "Budget"]]
```

### Join types

| Join | Keeps |
|---|---|
| `inner` | only rows that match in **both** tables |
| `left` | all rows from the left table, matched where possible |
| `right` | all rows from the right table, matched where possible |
| `outer` | all rows from **both** tables |


```python
inner = pd.merge(df, departments, on="Department", how="inner")
outer = pd.merge(df, departments, on="Department", how="outer")

print("inner shape:", inner.shape, "-> Finance dropped (no employees)")
print("outer shape:", outer.shape, "-> Finance kept, with NaN employee fields")
outer[outer["Department"] == "Finance"]
```

```python
# concat() -- stacking, NOT matching by key
new_hires_q1 = pd.DataFrame({
    "EmployeeID": [110, 111],
    "Name": ["Farida2", "Tamer"],
    "Department": ["IT", "Sales"],
    "City": ["Cairo", "Giza"],
    "Salary": [8000, 7500],
})
stacked = pd.concat([df[["EmployeeID", "Name", "Department", "City", "Salary"]], new_hires_q1], axis=0, ignore_index=True)
print(stacked.shape)
stacked.tail(4)
```

```python
# horizontal concat -- side by side (must share the same index alignment)
extra_info = pd.DataFrame({"Remote": [True, False, True, False, True, False, True, False, True]},
                           index=df.index)
pd.concat([df[["Name"]], extra_info], axis=1)
```

## 14. Applying Functions

`.apply()` runs a function over every value in a Series (or every row/column of a DataFrame) — for logic that built-in Pandas functions don't cover.

```python
df["Name"].apply(len)          # built-in function, applied element by element
```

```python
df["Salary"].apply(lambda x: x * 1.1)     # lambda -- an inline, throwaway function
```

```python
# apply() on a whole ROW (axis=1) -- useful when a calculation needs more than one column
df["Salary_plus_Bonus"] = df.apply(lambda row: row["Salary"] + row["Bonus"], axis=1)
df[["Name", "Salary", "Bonus", "Salary_plus_Bonus"]]
```

```python
# creating new boolean/derived features
df["HighEarner"] = df["Salary"] >= 8000
df["ExperienceAfter5Years"] = df["YearsExperience"] + 5
df[["Name", "Salary", "HighEarner", "YearsExperience", "ExperienceAfter5Years"]]
```

**✍️ Practice 14**
1. Use `.apply()` to create a column with the **length** of each `Department` name.
2. Use a lambda with `.apply()` to flag `True`/`False` whether `YearsExperience` is above the average.
3. Use row-wise `.apply(axis=1)` to compute `Salary / (YearsExperience + 1)` (a rough "salary per year" metric).


```python
# ✅ solution
print(df["Department"].apply(len))
avg_exp = df["YearsExperience"].mean()
print(df["YearsExperience"].apply(lambda x: x > avg_exp))
print(df.apply(lambda row: row["Salary"] / (row["YearsExperience"] + 1), axis=1).round(1))
```

## 15. Mapping Values

`.map()` replaces values using a dictionary — simpler and faster than `.apply()` when you're just doing a **lookup/substitution**, not a calculation.

```python
dept_full_name = {
    "IT": "Information Technology",
    "Sales": "Sales Department",
    "HR": "Human Resources"
}
df["Department"].map(dept_full_name)
```

```python
city_region = {"Cairo": "Greater Cairo", "Giza": "Greater Cairo", "Alex": "Delta"}
df["Region"] = df["City"].map(city_region)
df[["Name", "City", "Region"]]
```

### `apply()` vs `map()`

| | `apply()` | `map()` |
|---|---|---|
| Use for | calculations, custom logic | value replacement/lookup |
| Input | any function or lambda | usually a dict (or a Series) |
| Flexibility | high | simpler, faster for substitutions |
